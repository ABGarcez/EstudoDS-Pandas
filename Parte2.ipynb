*Manipulando Dados*

```python
import pandas as pd
df_inicial = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/base-de-dados/aluguel.csv', sep=";")
df_inicial
```

*Criando colunas para checar custo por mês e ano*

```python
df_inicial['valor_por_mes'] = df_inicial['Valor'] + df_inicial['Condominio']
df_inicial.head()
```

```python
df_inicial['valor_por_ano'] = df_inicial['valor_por_mes'] * 12 + df_inicial['IPTU']
df_inicial.head()
```

*Criando colunas categóricas*

```python
df_inicial['Descricao'] = df_inicial['Tipo'] + ' em ' + df_inicial['Bairro'] + \
                        ' com ' + df_inicial['Quartos'].astype(str) + ' quarto(s) ' + \
                        ' e ' + df_inicial['Vagas'].astype(str) + ' vaga(s) de garagem. '

df_inicial.head()
```

Nota: *Não é possivel concatenar strings com int, por isso utilizar o metodo astype para converter o int em string*

```python
df_inicial['possui_suite'] = df_inicial['Suites'].apply(lambda x: "sim" if x>0 else "não")
df_inicial.head()
```

*Salvando a base*

```python
df_inicial.to_csv('Filtro3.csv',index=False, sep=';')
```
