*Análise exploratória dos dados*

*Checando a media de aluguel por imovel*: 
    Utilizando a função groupby, podemos organizar a tabela a partir de uma coluna específica, porém sempre é utilizada em conjunto com uma função de agrupamento.

```python
dados.groupby('Tipo')['Valor'].mean()
```

Nota: *Em caso de multiplas colunas com tipos diferentes, aplicar o numeric_only=True no mean*

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Nota: *A adição do segundo par de [] fez o "Valor" virar um Dataframe, e a adição da função sort_values organizou os valores em ordem crescente*

*Plotando o gráfico pra melhor visualização*

```python
df_dados_valor = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_dados_valor.plot(kind='barh',figsize=(14,10), color='green')
```

*Removendo imoveis comerciais*

*Definindo quais imoveis são comerciais*

```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = [
        'Conjunto Comercial/Sala', 'Prédio Inteiro', 'Loja/Salão',
        'Galpão/Depósito/Armazém', 'Casa Comercial', 'Terreno Padrão', 
        'Box/Garagem', 'Loja Shopping/ Ct Comercial', 'Chácara',
        'Loteamento/Condomínio', 'Sítio', 'Pousada/Chalé', 'Hotel', 'Indústria'
]
```

Nota: *Metódo query serve para selecionar linhas do DataFrame, passando um critério de filtro, nesse caso, linhas onde os tipos comerciais não estejam presentes*

```python
dados.query('@imoveis_comerciais not in Tipo')
```

```python
imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')
```

*Plotando o gráfico novamente, desta vez apenas com imoveis residenciais*

```python
df_residencial_valor = imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_residencial_valor.plot(kind='barh',figsize=(14,10), color='green')
```

*Calculando porcentagem de tipo de imoveis*

```python
imoveis_residenciais.Tipo.value_counts(normalize=True)
```

Nota: *value_counts mostra a soma de valores do mesmo tipo. Para mostrar os valores em percentual, precisa adicionar o criterio normalize.*

```python
imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
```

Nota: *Método to_frame transforma a series em DataFrame*

```python
df_percentual_tipo = imoveis_residenciais.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipo.rename(columns={'proportion': 'Percentuais'},inplace=True)
df_percentual_tipo.plot(kind='bar',figsize=(14,10), color='green', 
                            xlabel='Tipos', ylabel='Percentual')
```

Nota: *Rename serve para renomear uma coluna de um DataFrame*
